## Задание 1

В предыдущем домашнем задании была реализована простейшная модель распространения вируса. Одним из недостатков той модели было то, что люди находились в плотную друг к другу и не передвигались. В этом задании мы добавим возможность перемещения людей. Как и прежде мы будем полагать

Человек может быть :

* Заражен. Носитель вируса. В конечном итоге либо становится имунным, либо мертвым.
* Не заражен. Может быть заражен носителем.
* Имунный. Человек, который переболел вирусом. Больше не может быть заражен и заражать других.
* Мертвый. Также не может быть заражен и заражать других.

Вирус имеет характиристики / параметры:
* Заразность - вероятность передать вирус от носителя к незараженному.
* Смертность - вероятность, что носитель вируса погибнет.
* Переносимость - вероятность, что носитель переболеет и получит иммунитет.

Однако теперь:

* Изначально пространство не должно быть полностью заполнено людьми
* Люди не могут наступать друг на друга и выходить за границы пространства
* Вирус может передаться человеку, расположенному на расстоянии двух клеток от носителя
* Погибший человек должен просто удаляться из пространства
* Люди могут ходить на одну клетку (вверх / вниз / вправо / влево или по диагонали)

Постройте данную модель и вкачестве результата выведите эволюцию систему и статистику о том как меняется число здоровых, заболевших, имунных и погибших людей.
