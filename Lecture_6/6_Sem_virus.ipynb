Задание 1
---
Реализовать простейшую модель распространения вируса.
Пространтсво дисктретизируется квадратной сеткой так, что в каждой ячейке оказывается только один человек. В модели люди неподвижны (не могут переходить из одной ячейки в другую). 

Человек может быть :

* Заражен. Носитель вируса. Может заражать ближайших соседей. В конечном итоге либо становится имунным, либо мертвым.
* Не заражен. Может быть заражен носителем.
* Имунный. Человек, который переболел вирусом. Больше не может быть заражен и заражать других.
* Мертвый. Также не может быть заражен и заражать других.

Вирус имеет характиристики / параметры:
* Заразность - вероятность передать вирус от носителя к незараженному.
* Смертность - вероятность, что носитель вируса погибнет.
* Переносимость - вероятность, что носитель переболеет и получит иммунитет.

Посмотреть эволюцию системы в стабильное/стационарное состояние (состояние системы стационарно при отуствии зараженных). Вывести как меняется число здоровых, заболевших, имунных и мертвых. 

В качетсве начального условия выберите четырех носителей вируса с заданным расположением. Параметры вируса:
* Заразность - 0.1
* Смертность - 0.01
* Переносимость - 0.05
